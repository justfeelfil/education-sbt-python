```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2, 4, 6, 7, 10]
y = [1, 3, 5]
plt.plot(x, [y for y in range(5)], 'bx-')
```

```python
x = np.linspace(-2,3, 500)
y = [0.5*j + 1 + np.random.normal(0, 0.2) for j in x]
plt.plot(x, y, 'gx')

x1 = np.linspace(-2, 3, 200)
y1 = [3 + np.random.normal(0, 0.07) for j in x1]
plt.plot(x1, y1, 'rx')
```

```python
second_x = np.concatenate((x,x1))
second_y = np.concatenate((y,y1))
```

```python
def linear(coeffs, x_i): 
    return coeffs[0]*x_i + coeffs[1]

def loss(coeffs):
    predictions = np.array([linear(coeffs, x_i) for x_i in second_x])
    ans = np.array(second_y)
    return sum((predictions - ans))

plt.plot(second_x, second_y, 'gx')    
#plt.plot(x, loss([0.5,1], 'gx')  
plt.plot(second_x, [linear([0.5,1],i) for i in second_x], 'b')
    
```

```python
loss([0.5, 1])
```

```python
from scipy.optimize import minimize
res =  minimize(loss, [1,10]).x
res
```
