# Text data visualization

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import wordcloud

from sklearn.datasets import fetch_20newsgroups 
```

## Wordcloud library

Wordcloud is a useful tool for represting text as a cloud of words. It is, in fact, a well-drawn counter of unique words. This allows you to see some basic features of a particular text. 

### Data loading

Based on such a wordclouds you can easilly see if is possible to distinguish between different text collections. Let's take a look at the example below: we load 2 different categories of new from the 20newsgroups dataset and create a worldclouds for them.

```python
space_news_data = fetch_20newsgroups(categories=['sci.space'], data_home='data/')
mideast_news_data = fetch_20newsgroups(categories=['talk.politics.mideast'], data_home='data/')
```

```python
print space_news_data.keys()
```

```python
print space_news_data['data'][0] 
```

```python
print mideast_news_data['data'][0] 
```

### Text processing

```python
space_news_text = ' '.join(space_news_data['data'])
mideast_news_text = ' '.join(mideast_news_data['data'])
```

### Wordcloud for sci.space category

```python
wordcloud_result = wordcloud.WordCloud(random_state=0).generate(space_news_text)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud_result, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Wordcloud for politics.mideast category 

```python
wordcloud_result = wordcloud.WordCloud(random_state=0).generate(mideast_news_text)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud_result, interpolation='bilinear')
plt.axis("off")
plt.show()
```
