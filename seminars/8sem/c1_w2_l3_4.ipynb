#### Image style

```python
plt.figure(figsize = (8, 8))

plt.subplot(1, 3, 1)
plt.imshow(digits.images[0])

plt.subplot(1, 3, 2)
plt.imshow(digits.images[0], cmap = 'summer', interpolation = 'gaussian')

plt.subplot(1, 3, 3)
plt.imshow(digits.images[0], cmap = 'gray', interpolation = 'gaussian')
```

#### Let's make a fire digits!

```python
plt.figure(figsize = (20, 8))

for plot_number, plot in enumerate(digits.images[:10]):
    plt.subplot(2, 5, plot_number + 1)
    plt.imshow(plot, cmap = 'hot', interpolation = 'gaussian')
    plt.title('digit: ' + str(digits.target[plot_number]))
```
